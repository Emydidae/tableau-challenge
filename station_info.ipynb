```python
# this is to make a csv with each station, its coordinates, number of trips to / from, and average trip duration
import pandas as pd
```

```python
trips1 = pd.read_csv('citi_aug_long.csv')
trips2 = pd.read_csv('citi_aug_short.csv')
trips = pd.concat([trips1, trips2], axis = 0)
trips.tail()
```

```python
start_grouped = trips.groupby(['start_station_name']).mean()
start_grouped.head()
```

```python
# get station name, latitude, longitude, average duration of trips leaving the station, average (peak) departure time
import math
start_df = start_grouped[['start_lat','start_lng','trip_duration', 'start_hr']].rename(columns={'start_lat': 'lat', 'start_lng': 'lng', 'trip_duration': 'avg_from_duration', 'start_hr': 'peak_leaving'})
start_df['peak_leaving'] = start_df['peak_leaving'].apply(math.floor)
start_df.head()
```

```python
start_grouped_count = trips.groupby(['start_station_name']).count()
start_grouped_count.head()
```

```python
# get number of trips leaving station & rank
start_df['trips_from'] = start_grouped_count['new_id']
start_df['from_rank'] = start_df['trips_from'].rank(ascending = False).astype('int')
start_df.head()
```

```python
end_grouped = trips.groupby(['end_station_name']).mean()
end_grouped.head()
```

```python
# get average duration of trips arriving and calculate average (peak) arrival time
end_df = end_grouped[['trip_duration']].rename(columns = {'trip_duration': 'avg_to_duration'})
end_df['peak_arriving'] = (end_grouped['start_hr'] + end_grouped['trip_duration'] / 60).apply(math.floor)
end_df.head()
```

```python
end_grouped_count = trips.groupby(['end_station_name']).count()
end_grouped_count.head()
```

```python
# get # of trips where the station was the destination & rank
end_df['trips_to'] = end_grouped_count['new_id']
end_df['to_rank'] = end_df['trips_to'].rank(ascending = False).astype('int')
end_df.head()
```

```python
# merge dfs
final_df = pd.merge(start_df, end_df, left_index = True, right_index = True)
final_df
```

```python
# get zip code, code from https://gis.stackexchange.com/questions/352961/converting-lat-lon-to-postal-code-using-python
import geopy
import time

zip_count = 0
zip_max = len(final_df)

def get_zipcode(df, geolocator, lat_field, lon_field):
    try:
        location = geolocator.reverse((df[lat_field], df[lon_field]))
    except:
        return None
    time.sleep(1)
    global zip_count
    zip_count += 1
    print(f'got {zip_count} / {zip_max}')
    try:
        return location.raw['address']
    except:
        print('wait that was wonky =;=;=;=;=;=;=;=;=')
        return None
geolocator = geopy.Nominatim(user_agent='citi_station_info')

zipcodes = final_df[['lat','lng']].apply(get_zipcode, axis=1, geolocator=geolocator, lat_field='lat', lon_field='lng')
```

```python
# get zipcodes for those that have them
import json
zipcodes_parsed = []
bad_list = []
for i in range(len(zipcodes)):
    try:
        zipcodes_parsed.append(zipcodes[i]['postcode'])
    except:
        print(i, end = ' ')
        print(zipcodes[i])
        bad_list.append(i)
        zipcodes_parsed.append(None)
print(len(zipcodes))
print(len(zipcodes_parsed))
```

```python
# trying again for the ones that failed
# basically I just kept running this and the cell above alternating until I had a few stubborn ones left
# I had already run the initial call 3 times which took like 40 minutes each so basically don't really care about this being elegant right now lol
for i in bad_list:
    print(i, end = ' - ')
    zipcodes[i] = get_zipcode(final_df.iloc[i,:],geolocator,'lat','lng')
    print(zipcodes[i])
```
